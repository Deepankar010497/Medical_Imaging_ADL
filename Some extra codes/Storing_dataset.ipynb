```python
import os
import re
import shutil
import numpy as np
import datetime
```

```python
# Renaming the files in sequencial order

def rename_all_files(file_path, cwd_case_class, count_copy_numbers, topic_name):
    
    ext = r".png"
    ext_2 = r".jpg"
    ext_3 = r"jpeg"

    cwd = cwd_case_class
    # noinspection PyBroadException
    try:
        os.chdir(file_path)
        local_count = 0
        files_list = os.listdir()
        #print("---", files_list)
        for file in files_list:
            #print(seconds)
            file_name = re.findall(r"([a-zA-Z0-9]+)\.", file)[0]
            file_ext = re.findall(r"(\.[a-zA-Z0-9]+)", file)[0]
            #print(file_ext)
            file_number = count_copy_numbers[local_count]
            #print(file_name)
            os.rename(file, f"{topic_name}_{str(file_number)}{file_ext}")
            local_count += 1
    except:
        print("Wrong directory  --- 1!")
    finally:
        os.chdir(cwd)
```
