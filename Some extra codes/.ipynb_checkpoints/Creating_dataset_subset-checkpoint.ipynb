```python
# To segregate files from the main dataset folder to my own dataset as required

# Make sure to change the files path accordingly in order to use the following code

# For copying and creating train subset dataset

import os
import re
import shutil

# Change the file_path, extension type and topics to your own choice and the run the program
train_file_path = r"E:\HP Deep Learning Internship\Dataset\train\XR_HAND"
test_file_path = r"E:\HP Deep Learning Internship\Dataset\valid\XR_HAND"

train_healthy_hand_image_folder = r"E:\HP Deep Learning Internship\Dataset\train\healthy_bones\hand"
train_fractured_hand_image_folder = r"E:\HP Deep Learning Internship\Dataset\train\fractured_bones\hand"
valid_healthy_hand_image_folder = r"E:\HP Deep Learning Internship\Dataset\valid\healthy_bones\hand"
valid_fractured_hand_image_folder = r"E:\HP Deep Learning Internship\Dataset\valid\fractured_bones\hand"

ext_1 = r".png"
ext_2 = r".jpg"
ext_3 = r".jpeg"

cwd_main = os.getcwd()

# noinspection PyBroadException
try:
    os.chdir(train_file_path)
    cwd_train_folder = os.getcwd()
    #print(cwd_train_folder)
    files_list = os.listdir()
    count = 0
    files_dict = {}
    sorted_files = []
    for file in files_list:          
        
        cwd_train_image_class = os.path.join(cwd_train_folder, file)
        #print(cwd_train_image_class)
        try:
            os.chdir(cwd_train_image_class)
            image_folders = os.listdir()
            #count_1 = 0
            for image_folder in image_folders:
                #print(image_folder)
                cwd_image_class = os.path.join(cwd_train_image_class, image_folder)
                class_name = re.findall(r"(positive|negative)", image_folder)[0].lower()
                if class_name == "positive":
                    #print("Abnormal/Fractured")
                    
                    # creating folder for each patient id
                    try:
                        path_1 = os.path.join(train_fractured_hand_image_folder, file)
                        os.mkdir(path_1)
                        #print('folder created')
                    except:
                        print("error in creating folder")
                    
                    # Creating folder for each class to store images
                    try:
                        path_2 = os.path.join(train_fractured_hand_image_folder, file, image_folder)
                        os.mkdir(path_2)
                        #print('folder created')
                    except:
                        print("error in creating folder")
                        
                    # Copying each image
                    try:
                        os.chdir(cwd_image_class)
                        images = os.listdir()
                        #count_2 = 0
                        for image in images:
                            #print(image)
                            cwd_image = os.path.join(cwd_image_class, image)
                            destination_path = os.path.join(train_fractured_hand_image_folder, file, image_folder)
                            shutil.copy(cwd_image, destination_path)
                    except:
                        print("Wrong Directory!")
                    finally:
                        os.chdir(cwd_image_class)
                    
                    
                    
                elif class_name == "negative":
                    #print("Normal/Healthy")
                    
                    # creating folder for each patient id
                    try:
                        path = os.path.join(train_healthy_hand_image_folder, file)
                        os.mkdir(path)
                        #print("folder created")
                    except:
                        print("error in creating folder")
                    
                    # Creating folder for each class to store images
                    try:
                        path = os.path.join(train_healthy_hand_image_folder, file, image_folder)
                        os.mkdir(path)
                        #print("folder created")
                    except:
                        print("error in creating folder")
                    
                    
                    # Copying each image
                    try:
                        os.chdir(cwd_image_class)
                        images = os.listdir()
                        #count_2 = 0
                        for image in images:
                            #print(image)
                            cwd_image = os.path.join(cwd_image_class, image)
                            destination_path = os.path.join(train_healthy_hand_image_folder, file, image_folder)
                            shutil.copy(cwd_image, destination_path)
                    except:
                        print("Wrong Directory!")
                    finally:
                        os.chdir(cwd_image_class)
                    
                
        except:
            print("Wrong directory!")
        finally:
            os.chdir(cwd_train_folder)
            
            
        
        #print(file, type(file))
        #print(os.path.join(cwd_train_folder, file))
        count += 1
        #print(os.getcwd())
            
except:
    print("Wrong directory!")
finally:
    os.chdir(cwd)

print(f"Number of files present are {count}")
```
